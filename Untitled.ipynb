```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import netCDF4 as netCDF
from IPython.display import Image
from scipy.fftpack import fft, ifft
from scipy import stats
from scipy.stats import norm
import cmath

import tabs
```

```python
def fft_wind(df,skip=1):
    dt = float((df.index[1]-df.index[0]).seconds * skip)
    N = len(df.index[::skip])
    T = N*dt
    dfrq = 1./T
    dfrq = 1./dt
    #detrend
    ut = df['42035: East [m/s] (air)'][::skip] - np.nanmean(df['42035: East [m/s] (air)'][::skip])
    vt = df['42035: North [m/s] (air)'][::skip] - np.nanmean(df['42035: North [m/s] (air)'][::skip])
    fn_u = fft(ut.notnull())*dt/T
    fn_v = fft(vt.notnull())*dt/T
    sf_u = np.absolute(fn_u)**2
    sf_v = np.absolute(fn_v)**2
    freq = np.asarray([ii*dfrq for ii in range(N)])
    return (fn_u, fn_v, sf_u, sf_v, freq, T, dt, dfrq)
```

```python
import pandas as pd
import xarray as xr
from datetime import datetime
```

```python
def make_frc(output='../tests/shelfstrat_frc.nc',
             uwind=0.0, vwind=5.0, Cd=1.5e-3, Rho0=1027.,
             ndays=30, dtw=1 / 24, Tramp=3.0, Tflat=3.):

    sustr0 = Cd * np.sqrt(uwind**2 + vwind**2) * uwind
    svstr0 = Cd * np.sqrt(uwind**2 + vwind**2) * vwind
    t = xr.DataArray(np.arange(0, ndays+dtw, dtw), dims=['sms_time'])
    if isinstance(vwind, float):
        ramp = (1.0 - np.exp(-t / Tramp)) * np.sin(t * 2.0 * np.pi)
        nt = int(Tflat / dtw)
        svstr = xr.DataArray(svstr0 * np.insert(ramp, 0, np.zeros(nt))[:-nt], dims=['sms_time'])
    else:
        svstr = xr.DataArray(svstr0[:int(ndays/dtw)+2], dims={'sms_time':t})
    if isinstance(uwind, float):
        sustr = sustr0 * (t - Tflat) / Tramp
        sustr.loc[t < Tflat] = 0
        sustr.loc[t > Tflat+Tramp] = sustr0
    else:
        sustr = xr.DataArray(sustr0[:int(ndays/dtw)+2], dims={'sms_time':t})
    # Create dataset

    ds = xr.Dataset({'sms_time': t, 'sustr': sustr, 'svstr': svstr})

    ds.attrs['Description'] = 'Initial conditions for ideal shelf'
    ds.attrs['Author'] = 'VRX'
    ds.attrs['Created'] = datetime.now().isoformat()
    ds.attrs['type'] = 'ROMS FRC file'
    ds['sms_time'].attrs['units'] = 'days'
    ds['sustr'].attrs['units'] = 'Newton meter-2'
    ds['svstr'].attrs['units'] = 'Newton meter-2'

    print('Writing netcdf FRC file: '+output)
    return ds
```

```python
url = 'http://pong.tamu.edu/tabswebsite/daily/42035_all'
df = pd.read_table(url, parse_dates=True, index_col=0) 
uw = df['East [m/s]'].loc['2010-08']
uw.replace(-999., np.nan, inplace=True)
uw = uw.interpolate()
vw = df['North [m/s]'].loc['2010-08']
vw.replace(-999., np.nan, inplace=True)
vw = vw.interpolate()
uw.to_xarray()
vw.to_xarray()
ds = xr.Dataset({'uw':uw, 'vw':vw})
ds.to_netcdf('winds.nc')
```
