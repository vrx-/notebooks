```python
import descarteslabs as dl
import matplotlib.pyplot as plt
import itertools
import cmocean
```

```python
roi = {
        "type": "Polygon",
        "coordinates": [
          [
            [
              -81.02691650390625,
              -3.5463175009986707
            ],
            [
              -79.66461181640625,
              -3.5463175009986707
            ],
            [
              -79.66461181640625,
              -1.9826364384297377
            ],
            [
              -81.02691650390625,
              -1.9826364384297377
            ],
            [
              -81.02691650390625,
              -3.5463175009986707
            ]
          ]
        ]
      }
```

```python
dl.metadata.products()
```

```python
dl.metadata.search(products=['modis:09:CREFL'])
```

```python
fc = dl.metadata.search(products=['modis:09:CREFL'], geom=roi, start_time='now-3M', limit=1000)
```

```python
fc
```

```python
items = itertools.groupby(
    [f for f in fc['features'] if f['properties']['sat_id']=='Aqua'], 
    key= lambda f: f['properties']['acquired'][:10]
    )
```

```python
items
```

```python
items = sorted([(k, list(v)) for (k,v) in items], key=lambda (k,v): k)
```

```python
items
```

```python
full_arr = None
fig, ax = plt.subplots(figsize=[16,16])
frames = []
for day, group in items:
    print(day)
    arr, meta = dl.raster.ndarray([g['id'] for g in group], srs='EPSG:32617',
                                 bands=['water_vapor'], cutline=roi, resolution=120, align_pixels=True)
    if full_arr is None:
        full_arr = arr.copy()
    
    mask = arr !=0
    full_arr[mask] = arr[mask]
    im = plt.imshow(full_arr)
    t = plt.text(30, full_arr.shape[0]-40, day, color='red', size='large')
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    fname = "%s_demo_gye.png" % day
    out = plt.savefig(fname, bbox_inches='tight')
    t.remove()
    frames.append(fname)
del fig, ax
```
