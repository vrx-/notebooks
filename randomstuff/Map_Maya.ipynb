```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

from cartopy.io.img_tiles import StamenTerrain
%matplotlib inline
```

```python
# Grid projection settings
lat_top=30.
lat_bot=27.5
lon_left=-97.
lon_rigth=-92.

lon_0 = (lon_left+lon_rigth)*0.5

extent=[lon_left, lon_rigth, lat_bot, lat_top]

p = ccrs.PlateCarree(central_longitude=lon_0) #Projection

res="10m"
```

```python
tiler = StamenTerrain()
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, projection=p)
ax.set_extent(extent)
ax.add_image(tiler, 8)
ax.coastlines(res)
```

```python
fig = plt.figure(figsize=(10,5)) #tamaño de la figura
ax = fig.add_subplot(111, projection=p) #p es la proyeccion
ax.set_extent(extent) #extent es la delimitacion del mapa
ax.coastlines(res) # la resolucion de la costa (10 m)
ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                            name='land', facecolor='grey',
                                            scale=res)) #add land
ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                            name='rivers_lake_centerlines',
                                            edgecolor='blue', facecolor='none',
                                            scale=res)) #add rivers

#la grilla:
gl = ax.gridlines(xlocs=np.arange(lon_left, lon_rigth+1), ylocs=np.arange(lat_bot, lat_top+1),
                  draw_labels=True, linewidth=0.5, color='k', linestyle=':')
gl.xlabels_top = gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER

#los puntos: x, y
ax.plot(-95, 28.5, 'or', transform=ccrs.PlateCarree())
ax.plot(-93.7, 28.1, 'xb', transform=ccrs.PlateCarree())
ax.plot(-94.2, 28.7, '*g', transform=ccrs.PlateCarree())
```

```python
fig = plt.figure(figsize=(10,5)) #tamaño de la figura
ax = fig.add_subplot(111, projection=p) #p es la proyeccion
ax.set_extent(extent) #extent es la delimitacion del mapa
ax.coastlines(res) # la resolucion de la costa (10 m)
ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                            name='land', facecolor=cfeature.COLORS['land'],
                                            scale=res)) #add land
ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                            name='rivers_lake_centerlines',
                                            edgecolor=cfeature.COLORS['water'], facecolor='none',
                                            scale=res)) #add rivers
ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                            name='lakes',
                                            scale=res,
                                           edgecolor='none', facecolor=cfeature.COLORS['water']))

#la grilla:
gl = ax.gridlines(xlocs=np.arange(lon_left, lon_rigth+1), ylocs=np.arange(lat_bot, lat_top+1),
                  draw_labels=True, linewidth=0.5, color='k', linestyle=':')
gl.xlabels_top = gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER

#los puntos: x, y
ax.plot(-95, 28.5, 'or', transform=ccrs.PlateCarree())
ax.text(-94.95, 28.5, 'Gay Catnip', transform=ccrs.PlateCarree())
ax.plot(-93.7, 28.1, 'xb', transform=ccrs.PlateCarree())
ax.text(-93.65, 28.1, 'Sandia', transform=ccrs.PlateCarree())
ax.plot(-94.2, 28.7, '*g', transform=ccrs.PlateCarree())
ax.text(-94.15, 28.7, 'Burnt Blowfish', transform=ccrs.PlateCarree())
ax.plot(-93.8, 29.0, 'pc', transform=ccrs.PlateCarree())
ax.text(-93.75, 29.0, 'Bind AF', transform=ccrs.PlateCarree())

ax.set_title('The Queer as Folk distribution map', fontsize=18)
```
