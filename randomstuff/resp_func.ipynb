```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
def SOD(btm_tmp, btm_ox):
    return 6. * 2.**(.1 * btm_tmp) * (1. - np.exp(- .033 * btm_ox))
```

```python
temps = np.linspace(30, 20, 5)
ox = np.linspace(0,200.,100)
```

```python
temps = np.linspace(30, 20, 5)
ox = np.linspace(0,200.,100)
fig, ax = plt.subplots(figsize=(12,8))
alpha = 1.
for t in temps:
    resp = SOD(t, ox)
    ax.plot(ox, resp, 'r', alpha=alpha)
    alpha-=.2
    ax.text(175, resp[75]-1.5, str(t)+'$^\circ C$', color='r')
ax.set_xlabel('DO concentration $[kg/m^3]$', fontsize=18)
ax.set_ylabel('SOD $[kg/m^2 d]$', fontsize=18 )
ax.tick_params(axis='both', labelsize=16)
# 'Temperature $[^\circ C]$'
```

```python
#ox,temp
resp = np.zeros((200,200))
temps = np.linspace(0, 40, 200)
ox = np.linspace(0,200.,200)
for i,t in enumerate(temps):
    resp[:,i] = SOD(t, ox)
    
fig, ax = plt.subplots(figsize=(12,8))
ax.pcolor(ox, temps, resp, cmap='Reds' )
# ax.semilogy()
# ax.loglog()
ax.set_xlim(0., 200)
ax.set_ylim(0., 40)
```
